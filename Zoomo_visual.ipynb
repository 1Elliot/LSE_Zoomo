```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
accel1=pd.read_csv('Accelerator1_cleaned.csv')
```

```python
accel1.head
```

```python
accel1.drop(['X', 'Y', 'Z'], axis=1)
```

```python
# Chart for Forward/reverse over time
sns.pointplot(x='Time', y='X_f', data = accel1)
```

```python
# Chart for Up/Down over time
sns.pointplot(x='Time', y='Y_f', data = accel1)
```

```python
# Chart for Left/Right over time
sns.pointplot(x='Time', y='Z_f', data = accel1)
```
